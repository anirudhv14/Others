```python
def swap_case(s):
    res = ''
    for i in s:
        if(i != ' '):
            if(str.islower(i)):
                res += i.upper()
            elif(str.isupper(i)):
                res += i.lower()
            else:
                res += i
        else:
            res += ' '
    return res

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
def mutate_string(string, position, character):
    res = ''
    for i in range(len(string)):
        if(i == position):
            res += character
        else:
            res += string[i]
    return res

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

```python
def count_substring(string, sub_string):
    ctr = 0
    for i in range(0,len(string)-len(sub_string)+1):
        if(string[i:i+len(sub_string)] == sub_string):
            ctr += 1
    return ctr

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    count = count_substring(string, sub_string)
    print(count)
```

```python
def checkAlphaNum(s):
    for i in s:
        if(i.isalnum()):
            return True
    return False
def checkAlpha(s):
    for i in s:
        if(i.isalpha()):
            return True
    return False
def checkDigit(s):
    for i in s:
        if(i.isdigit()):
            return True
    return False
def checkLower(s):
    for i in s:
        if(i.islower()):
            return True
    return False
def checkUpper(s):
    for i in s:
        if(i.isupper()):
            return True
    return False
        

if __name__ == '__main__':
    s = input()
    print(checkAlphaNum(s))
    print(checkAlpha(s))
    print(checkDigit(s))
    print(checkLower(s))
    print(checkUpper(s))
```

```python
def minion_game(string):
    s1 = s2 =string
    ssc = []
    ksv = []
    for ch in string:
        if(ch=='A' or ch=='a' or ch=='E' or ch =='e' or ch=='I'
           or ch=='i' or ch=='O' or ch=='o' or ch=='U' or ch=='u'):
            if(ch not in ksv):
                ksv.append(ch)
        else:
            if(ch not in ssc):
                ssc.append(ch)
    finResStuart = []
    finResKevin = []
    for ctr in range(1,len(s1)+1):
        q = []
        for i in ssc:
            for j in range(0,len(s1)):
                if(s1[j:j+ctr].startswith(i)):
                    if(len(s1[j:j+ctr]) == ctr):
                        if(s1[j:j+ctr] not in q):
                            q.append(s1[j:j+ctr])
        if(len(q) != 0):
            finResStuart.append(q.copy())
        
    for ctr in range(1,len(s2)+1):
        q = []
        for i in ksv:
            for j in range(0,len(s2)):
                if(s2[j:j+ctr].startswith(i)):
                    if(len(s2[j:j+ctr]) == ctr):
                        if(s2[j:j+ctr] not in q):
                            q.append(s2[j:j+ctr])
        if(len(q) != 0):
            finResKevin.append(q.copy())
    stuart = {}
    kevin = {}
    for i in finResStuart:
        for j in i:
            stuart[j] = s1.count(j)
    for i in finResKevin:
        for j in i:
            kevin[j] = s2.count(j)
    
    if(sum(kevin.values()) > sum(stuart.values())):
        print(f"Kevin {sum(kevin.values())}")
    elif(sum(kevin.values()) < sum(stuart.values())):
        print(f"Stuart {sum(stuart.values())}")
            
if __name__ == '__main__':
    s = input()
    minion_game(s)
```
